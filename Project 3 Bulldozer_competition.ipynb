```python
df=pd.read_csv("TrainAndValid.csv",parse_dates=(["saledate"]),low_memory=False)
```

```python
df
```

```python
df.sort_values(by=("saledate"),ascending=True,inplace=True)
```

```python
df_tmp=df.copy()
```

```python
df_tmp
```

```python
df_tmp["saleYear"] = df_tmp.saledate.dt.year
```

```python
# This will turn all of the string values into category values
for label, content in df_tmp.items():
    if pd.api.types.is_string_dtype(content):
        df_tmp[label] = content.astype("category").cat.as_ordered()
```

```python
# Split data into training and validation
df_val = df_tmp[df_tmp.saleYear == 2012]
df_train = df_tmp[df_tmp.saleYear != 2012]
x_train, y_train = df_train.drop("SalePrice", axis=1), df_train.SalePrice
x_valid, y_valid = df_val.drop("SalePrice", axis=1), df_val.SalePrice
```

```python
for label, content in df_tmp.items():
    if pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
x_train
```

```python
x_valid
```

```python
# Fill numeric rows with the median
for label, content in x_train.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            # Add a binary column which tells if the data was missing our not
           # x_train[label+"_is_missing"] = pd.isnull(content)
            # Fill missing numeric values with median since it's more robust than the mean
            x_train[label] = content.fillna(content.median())
            # Turn categorical variables into numbers
    if not pd.api.types.is_numeric_dtype(content):
        # Add binary column to inidicate whether sample had missing value
       # df_tmp[label+"_is_missing"] = pd.isnull(content)
        # We add the +1 because pandas encodes missing categories as -1
        x_train[label] = pd.Categorical(content).codes+1        
```

```python
x_train
```

```python
# Fill numeric rows with the median
for label, content in x_valid.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            # Add a binary column which tells if the data was missing our not
           # x_train[label+"_is_missing"] = pd.isnull(content)
            # Fill missing numeric values with median since it's more robust than the mean
            x_valid[label] = content.fillna(content.median())
            # Turn categorical variables into numbers
    if not pd.api.types.is_numeric_dtype(content):
        # Add binary column to inidicate whether sample had missing value
       # df_tmp[label+"_is_missing"] = pd.isnull(content)
        # We add the +1 because pandas encodes missing categories as -1
        x_valid[label] = pd.Categorical(content).codes+1        
```

```python
x_valid
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_jobs=-1,random_state=42)
%time
model.fit(x_train,y_train)
```

```python
#creating evalution function 
from sklearn.metrics import mean_squared_log_error,r2_score,mean_absolute_error
def rmsle(y_test,y_preds):
    return(np.sqrt(mean_squared_log_error(y_test,y_preds)))
```

```python
def score(model):
    train_preds=model.predict(x_train)
    val_preds=model.predict(x_valid)
    score={"Training MAE":mean_absolute_error(y_train,train_preds),"Validation MAE":mean_absolute_error(y_valid,val_preds),
          "Training R2":r2_score(y_train,train_preds),"Validation R2":r2_score(y_valid,val_preds),
           "Training RMSLE":rmsle(y_train,train_preds),"Validation RMSLE":rmsle(y_valid,val_preds)}
    return score
```

```python
score(model)
```

```python
model2=RandomForestRegressor(n_jobs=-1,random_state=42,max_samples=10000)
%time
model2.fit(x_train,y_train)
score(model2)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf_grid={"n_estimators": np.arange(10,100,10),"max_depth": [None,3,5,10],"min_samples_split": np.arange(2,20,2),"min_samples_leaf": np.arange(1,20,2),"max_features":[0.5,1,"sqrt","auto"]}
```

```python
model3=RandomizedSearchCV(RandomForestRegressor(n_jobs=-1,random_state=42),param_distributions=rf_grid,n_iter=2,cv=5,verbose=True)
model3.fit(x_train,y_train)
```

```python
model3.best_params_
```

```python
score(model3)
```
